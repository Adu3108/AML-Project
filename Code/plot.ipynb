```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df_mat = pd.read_csv("./Results/FashionMNIST_MMD_MAP_mat_accuracy.csv", index_col=False)
# df_vae = pd.read_csv("./Results/MNIST_MMD_MAP.csv", index_col=False)
df_mmd = pd.read_csv("./Results/FashionMNIST_MMD_MAP.csv")

# df_vae.drop("Unnamed: 0", axis=1, inplace=True)
df_mat.drop("Unnamed: 0", axis=1, inplace=True)
df_mmd.drop("Unnamed: 0", axis=1, inplace=True)

df_mat.drop(df_mat.index[1], inplace=True)
```

```python
plt.plot(df_mmd["Embedding Dimension"], df_mat["Mean Accuracy"], color='orange', marker='o', label='MRL')
plt.plot(df_mmd["Embedding Dimension"], df_mmd["Mean Accuracy"], color='blue', marker='o', label='MMD')
plt.xlabel("Embedding Dimension")
plt.ylabel("Mean Accuracy")
plt.legend()
plt.savefig("./Presentation/Images/FashionMNIST_MMD_MAP_MRL.png")
```
